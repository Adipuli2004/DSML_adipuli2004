```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

## functional method

```python
plt.plot(x,y) #making the plot
plt.xlabel("x axis") # x axis label
plt.ylabel("Y axis") # y axis label
plt.title("first plot") # title label
```

## multiplots

```python
plt.subplot(1,2,1)#sub plot take number of row columns and plot number as arg
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object oriented method 

```python
fig=plt.figure()#making figure element
axes=fig.add_axes([0.1,0.1,0.8,0.8])#adding axes
axes.set_xlabel("X axis")#setting axes
axes.set_ylabel("Y axis")
axes.set_title("OO method")#title
axes.plot(x,y)#plotting
```

## Plots inside plots

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])#first axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3])#second axes
axes1.plot(x,y)#plotting on first axes
axes1.set_title('bigger plot')
axes2.plot(y,x)#plotting on second axes
axes.set_title('Smaller plot')
```

## Subplots

```python
fig,axes=plt.subplots(nrows=1,ncols=2)#adds axes on its own
axes#it is a list of axes 
#you can iterate through it using for loop or simply calling
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_xlabel("x1")
axes[1].set_xlabel("x2")
plt.tight_layout()#fixes all the overlapping 
plt.show()
```

## You can traverse through axes

```python
fig,axes=plt.subplots(nrows=3,ncols=2)
for i in axes:
    for j in i:
        j.plot(x,y,'r')
plt.tight_layout()#fixes all the overlapping 
```
