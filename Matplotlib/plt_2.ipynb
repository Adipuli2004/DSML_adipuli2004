```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,101)
y=x**2
```

## parameters

```python
fig=plt.figure(figsize=(5,1),dpi=500)#fig size takes size in inches
#dpi makes the plot look cleaner
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.tight_layout()
```

## saving plots

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=1000)
#fig size,dpi also works in subplot
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'b')
plt.tight_layout()
plt.show()
fig.savefig("myplot.png")#saves your plot 
fig.savefig("mypot2.jpg",dpi=50)
```

## legends

```python
fig=plt.figure(dpi=500)
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,'r',label="X squared")#labeling each plot 
axes.plot(x,x**3,'b',label="X cubed")
axes.legend()#makes a lengend with the given colors
#the legend might appear in some intersecting location 
#you can specify location using strings and location codes given in documentation
# best to use location code 0 for best
axes.legend(loc=0)
# you can also use a tuple of position (0.1,0.1,)
```

```python
x=np.linspace(0,5,11)
y=x**2
z=np.ones(11)*60
a=np.ones(11)*40
```

## line width styles color 

```python
fig=plt.figure(dpi=500)
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='black')#color can accept simple colors as strings
#it also accepts RGB hex codes as parameter
axes.plot(x,x**3,color='#FF0088',lw=0.5)
axes.plot(x,x,color='red',linewidth=3,alpha=0.5)
#line width multiplied by default line width
#alpha controls transparency percentage
#line width can also be typed as lw
#line style is used to change line type
axes.plot(x,z,color='blue',lw=2,linestyle='dotted')
axes.plot(x,a,color='yellow',lw=2,linestyle='--')
# you can see a lot line styles in notebook
# line style can also be shortened by using ls
```

## Markers

```python
fig=plt.figure(dpi=500)
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,'violet',lw=0.8,marker='+',markersize=10)
#marker type can be specified using strings ans also integers 
#look at documentation for more styles
axes.plot(x,x**3,lw=0.8,marker='o',markerfacecolor='red',markeredgewidth=2,markeredgecolor='black')
#you can specify marker properties 
plt.tight_layout()
```

## Limits

```python
x=np.linspace(0,10,101)
y=(x-5)**2
fig=plt.figure(dpi=500)
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,'black')
axes.set_xlim([0,10])#limit of axes
axes.set_ylim([0,25])
```
