```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.datasets import fetch_california_housing
```

```python
cal=fetch_california_housing()
```

```python
cal
```

```python
X=pd.DataFrame(cal['data'],columns=cal['feature_names'])
```

```python
X.info()
```

```python
X.head()
```

```python
X.describe()
```

```python
y=pd.DataFrame(cal['target'],columns=cal['target_names'])
```

```python
y.head()
```

```python
y.info()
```

```python
y.describe()
```

```python
X.drop(['Latitude','Longitude'],axis=1,inplace=True)
```

```python
X.info()
```

```python
sns.pairplot(X)
```

```python
sns.scatterplot(x=X['MedInc'],y=y['MedHouseVal'])
```

```python
sns.displot(y['MedHouseVal'])
```

```python
lm=LinearRegression()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
lm.fit(X_train,y_train)
```

```python
predictions=lm.predict(X_test)
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.mean_squared_error(y_test,predictions)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```
